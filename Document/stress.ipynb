# Stretch energy calculation and its diravatives

```python
import sympy as sp
from sympy import symbols, integrate, expand
```

```python
xs = symbols('x_{1:13}', real=True)
x = sp.Matrix(xs)
x
```

```python
Dm_tmp = symbols('D_m11 D_m21 D_m31 D_m12 D_m22 D_m32 D_m13 D_m23 D_m33', real=True)
Dm_tmp = sp.Matrix(Dm_tmp)
Dm = sp.Matrix.zeros(3,3)
Dm[0:3,0] = Dm_tmp[0:3]
Dm[0:3,1] = Dm_tmp[3:6]
Dm[0:3,2] = Dm_tmp[6:9]
Dm
```

```python
x1 = sp.Matrix(x[0:3, 0])
x2 = sp.Matrix(x[3:6, 0])
x3 = sp.Matrix(x[6:9, 0])
x4 = sp.Matrix(x[9:12, 0])
```

```python
Ds = sp.Matrix.zeros(3,3)
Ds[0:3,0] = x2 - x1
Ds[0:3,1] = x3 - x1
Ds[0:3,2] = x4 - x1

F = Ds * Dm
F
```

```python
F_vec = F.vec()
F_vec
```

```python
dFdx = (F_vec).jacobian(x)
dFdx
```
